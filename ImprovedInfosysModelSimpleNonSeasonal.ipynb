```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

infosys = pd.read_csv(r'C:\Users\MY\Downloads\Datasets\Nifty50\INFY.csv')

infosys.head()
```

```python
from datetime import datetime

month = []
year = []

Date = infosys['Date']

for d in Date:
    d1 = datetime.strptime(d, '%Y-%m-%d')
    m = d1.month
    y = d1.year
    month.append(m)
    year.append(y)
```

```python
infosys['month'] = month
infosys['year'] = year
```

```python
infosys.head()
```

```python
infosys['monthly_avg'] = infosys.groupby(['year', 'month']).Close.transform('mean')
```

```python
infosys.head()
```

```python
infosys.drop_duplicates(subset = ['monthly_avg'], keep = 'first', inplace = True)
```

```python
import pmdarima as pm

model = pm.auto_arima(infosys['monthly_avg'], start_p = 0, start_q = 0,
                      test='adf',       # use adftest to find optimal 'd'
                      max_p = 3, max_q = 3, # maximum p and q
                      m = 1,              # frequency of series
                      d = None,           # let model determine 'd'
                      seasonal = False,   # No Seasonality
                      D = 0,
                      trace = True,
                      error_action = 'ignore',  
                      suppress_warnings = True, 
                      stepwise = True)

print(model.summary())
```

```python
n_periods = 24

fitted, confidence = model.predict(n_periods = n_periods, return_conf_int = True)
index_of_fc = pd.date_range(infosys['monthly_avg'].index[-1], periods = n_periods, freq = 'MS')


fitted_series = pd.Series(fitted, index = index_of_fc)
lower_series = pd.Series(confidence[ : , 0], index = index_of_fc)
upper_series = pd.Series(confidence[ : , 1], index = index_of_fc)

plt.figure(figsize = (14, 6))
plt.plot(infosys['monthly_avg'], label = 'Closing')
plt.plot(fitted_series, color = 'darkgreen', label = 'forecasted')
plt.fill_between(lower_series.index,
                lower_series,
                upper_series,
                color = 'k',
                alpha = 0.15)

plt.title("SARIMA - Final Forecast of Infosys Data")
plt.legend()
plt.show()
```
