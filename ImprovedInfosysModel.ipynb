```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

infosys = pd.read_csv(r'C:\Users\MY\Downloads\Datasets\Nifty50\INFY.csv')

infosys.head()
```

```python
infosys.info()
```

```python
from datetime import datetime

month = []
year = []

Date = infosys['Date']

for d in Date:
    d1 = datetime.strptime(d, '%Y-%m-%d')
    m = d1.month
    y = d1.year
    month.append(m)
    year.append(y)
```

```python
infosys['month'] = month
infosys['year'] = year
```

```python
infosys.head()
```

```python
infosys['monthly_avg'] = infosys.groupby(['year', 'month']).Close.transform('mean')
```

```python
infosys.head()
```

```python
infosys.drop_duplicates(subset = ['monthly_avg'], keep = 'first', inplace = True)
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

result = seasonal_decompose(infosys['monthly_avg'], extrapolate_trend = 'freq', period = 12, model = 'additive')
result.plot()
```

```python
 decomposition_trend = result.trend
 ax = decomposition_trend.plot(figsize=(14,2))
 ax.set_xlabel('Date')
 ax.set_ylabel('Trend of time series')
 ax.set_title('Trend values of the time series')
 plt.show()
```

```python
detrended = infosys['monthly_avg'].values - result.trend

detrended_df = pd.DataFrame(detrended)
detrended_df.head()
detrended_df.fillna(0.0, inplace = True)
```

```python
detrended_df.shape
```

```python
detrended_df.plot()
```

```python
import pmdarima as pm

model = pm.auto_arima(detrended_df['trend'], start_p=0, start_q=0,
                      test='adf',       # use adftest to find optimal 'd'
                      max_p=3, max_q=3, # maximum p and q
                      m=1,              # frequency of series
                      d=None,           # let model determine 'd'
                      seasonal=False,   # No Seasonality
                      D=0, 
                      trace=True,
                      error_action='ignore',  
                      suppress_warnings=True, 
                      stepwise=True)

print(model.summary())
```

```python
n_periods = 24

fitted1 = model.predict(n_periods=n_periods)
index_of_fc = pd.date_range(infosys['monthly_avg'].index[-1], periods = n_periods, freq = 'MS')


fitted_series1 = pd.Series(fitted1, index = index_of_fc)

plt.figure(figsize = (14, 6))
plt.plot(infosys['monthly_avg'], label = 'Closing')
plt.plot(fitted_series1, color = 'darkgreen', label = 'forecasted')

plt.title("SARIMA - Final Forecast of Infosys Data")
plt.legend()
plt.show()
```
